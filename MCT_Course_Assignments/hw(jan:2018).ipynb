```python
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
#y = m x + c with m and c
m = float(input('enter the slope of the line '))
c = float(input('enter the constant '))
x = np.arange(0,10,0.1)
y = m*x+c

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y,label="y=mx+c")
plt.legend()
plt.show()
```

```python
#y = x^2, y = a x^2 + b x +c, with different a, b, c
a = float(input('enter first constand of quadratic equation, a : '))
b = float(input('enter second constand of quadratic equation, b : '))
c = float(input('enter third constand of quadratic equation, c : '))
x = np.arange(-10,10,0.1)
y1 = x**2 
y2 = a*(x**2)+b*x+c

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y1,label="y1=x^2")
plt.plot(x,y2,label="y2=ax^2+bx+c")
plt.legend()
plt.show()
```

```python
#y = sin x, y = sin 3 x, y = a sinx x + b cos x
a = float(input('enter first constand of the trignometric equation, a : '))
b = float(input('enter second constand of the trignometric equation, b : '))
x = np.arange(-10,10,0.1)

y1 = np.sin(x)
y2 = np.sin(x)*a*x + np.cos(x)*b
y3 = np.sin(3*x)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y1,label="y1=sinx")
plt.plot(x,y2,label="y2=ax(sinx)+b(cosx)")
plt.plot(x,y3,label="y3=sin3x")
plt.legend()
plt.show()
```

```python
#y = exp x, y = exp 2x, and y = exp x/2
x = np.arange(-1,1,0.001)
y = np.exp(x)
z = np.exp(2*x)
w = np.exp(x/2)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y,label="e^x")
plt.plot(x,z,label="e^2x")
plt.plot(x,w,label="e^x/2")
plt.legend()
plt.show()
```

```python
#y = ln x, y = log10 x
x = np.arange(0.1,100,0.1)
y = np.log(x)
z = np.log10(10*x)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y, label="ln(x)")
plt.plot(x,z, label="log(10x)")
plt.legend()
plt.show()
```

```python
#y = ln (1 + x) for 0 < x < 1
x=np.arange(0,1,0.001)
y=np.log(1+x)

plt.xlabel("range 0<x<1")
plt.ylabel("ln(1+x)")
plt.plot(x,y)
plt.show()
```
