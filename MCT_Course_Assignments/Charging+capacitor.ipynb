```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import random
```

```python
t=np.linspace(0,8,100)
R=float(input("Enter R: "))
C=float(input("Enter C: "))
V=float(input("Enter V: "))
Q=C*V*(1-np.exp(-(t/(R*C))))
plt.figure()
plt.plot(t,Q,label="Charging of capacitor")
plt.xlabel('t')
plt.ylabel('Q(t)')
plt.legend()
```
