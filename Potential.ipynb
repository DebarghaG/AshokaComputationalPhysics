```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.arange(-7,7,0.01)
a = float(input("m = "))
b = float(input("L = "))
c = 10 #m/s^2
y = a*b*c*(1-np.cos(x))
plt.plot(x,y, label = " Potetial of Pendulum ")
plt.xlabel("Theta",size = 20)
plt.ylabel("Potential",size = 20)
plt.legend()
plt.show()
```
