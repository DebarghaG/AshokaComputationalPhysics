```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
a = float(input("Acceleration = "))
c = float(input ("Initial Velocity = "))
x = np.arange(-10,10,0.1)
y = a*x+c
plt.plot(x,y, label = "$V-T$ graph")
plt.xlabel("Time", size = 20)
plt.ylabel("Velocity", size = 20)
plt.legend()
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
k = float(input("Spring constant = "))
y = 2*np.pi*((x/k)**0.5)
plt.plot(x,y, label="T-M graph")
plt.xlabel("rt.Mass", size = 20)
plt.ylabel("Time", size = 20)
plt.legend()
plt.show()
```

```python
x = np.arange(0,10,0.1)
k = float(input("Spring Constant = "))
y = (4*((np.pi)**2)*x)/k
plt.plot(x,y, label = "T^2 - M graph")
plt.xlabel("Mass", size = 20)
plt.ylabel("T^2", size = 20)
plt.legend()
plt.show()
```

```python
a = float(input("C = "))
b = float(input("R = "))
c = float(input("V = "))
x = np.arange(0,a*b,0.01)
y = b*(1-np.exp(-x/(a*b)))
plt.plot(x,y)
plt.xlabel("time", size = 20)
plt.ylabel("Voltage",size = 20)
plt.legend()
plt.show()
```

```python
x = np.arange (-10,10,0.1)
A = float(input("Enter A : "))
b = float(input("Enter b : "))
k = float(input("Enter X0 : "))
y = A*np.exp(-(b* (x-k)**2))
plt.plot(x,y)
plt.xlabel("X", size = 20)
plt.ylabel("Y", size = 20)
plt.legend()
plt.show()
```

```python
x = np.arange (-10,10,0.1)
A = float(input("Enter A : "))
b = float(input("Enter b : "))
k = float(input("Enter X0 : "))
y = A*np.exp(-(b* (x-k)**2))
m = A*np.exp(-(b* x**2))
plt.plot(x,y)
plt.plot(x,m)
plt.xlabel("X", size = 20)
plt.ylabel("Y", size = 20)
plt.legend()
plt.show()
```

```python
x = np.arange (-10,10,0.1)
A = float(input("Enter A : "))
b = float(input("Enter b : "))
c = float(input("Enter c : "))
k = float(input("Enter X0 : "))
y = A*np.exp((-(b* (x-k)**2))+ c*x)
plt.plot(x,y)
plt.xlabel("X", size = 20)
plt.ylabel("Y", size = 20)
plt.legend()
plt.show()
```

```python
x = np.arange (-10,10,0.1)
n = float(input("Enter n : "))
y = (x**n)*np.exp(-x)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,1,0.01)
y = 1/(1-x)
z = 1 + x + x**2
m = 1 + x + x**2 + x**3
n = 1 + x + x**2 + x**3 + x**4
plt.plot(x,y, label = "1/(1+x)")
plt.plot(x,z, label = "Approx 1")
plt.plot(x,m, label = "Approx 2")
plt.plot(x,n, label = "Approx 3")
plt.legend()
plt.show()
```
