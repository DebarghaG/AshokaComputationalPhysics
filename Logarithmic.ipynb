```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=np.arange(0.01,5,0.01)
y=np.log(x)
z=np.log10(x)
plt.plot(x,y, label = "lnx", linewidth = "2")
plt.plot(x,z, label = "logx", linewidth = "2")
plt.xlabel("$x$ axis", size = 20)
plt.ylabel("$y$ axis", size = 20)
plt.legend()
plt.show()
```
