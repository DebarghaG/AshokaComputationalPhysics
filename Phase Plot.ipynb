```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
a = float(input("W = "))
b = float(input("Q(0) = "))
x = np.arange(-10,10,0.001)
y = -(a)*b*(np.sqrt(1-((x/b)**2)))
z = (a)*b*(np.sqrt(1-((x/b)**2)))
plt.plot(x,y)
plt.plot(x,z)
plt.xlabel("Q(t)", size = 20)
plt.ylabel("Q'(t) ", size = 20)
plt.legend()
plt.show()
```
